```python
import pandas as pd
import numpy as np
import seaborn as snb
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error
```

```python
data= pd.read_excel('Data_Train.xlsx')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['Engine'][0]
```

```python
data=data.drop_duplicates()
```

```python
train=data.copy()
train.drop(['Name','Location'],axis=1)
```

```python
train.dtypes.value_counts()
```

```python
import datetime as dt
train['Year'] = train['Year'].apply(lambda year: dt.datetime.now().year - year)
```

```python
train['Year']
```

```python
train['Mileage']=train['Mileage'].str.rstrip('kmplkm/kg').astype('float')
```

```python
train['Mileage']
```

```python
train['Power']=train['Power'].str.rstrip('bhp')
```

```python
train['Power']
```

```python
train['Power']= pd.to_numeric(train['Power'],errors='coerce')
```

```python
train['Power']
```

```python
train.isnull().sum()
```

```python
train['Mileage'].fillna(train['Mileage'].astype('float').mean(), inplace = True)
train['Power'].fillna(train['Power'].astype('float').mean(), inplace = True)
train['Engine'].fillna(train['Engine'].astype('float').mean(), inplace = True)
train['Seats'].fillna(train['Seats'].astype('float').mean(), inplace = True)
```

```python
train.isnull().sum()
```

```python
train.columns
```

```python
train = pd.get_dummies(train,columns=['Fuel_Type', 'Transmission', 'Owner_Type'],drop_first = True)
```

```python
train
```

```python
train.drop('Location',axis=1)
```

```python
train.drop('Name',axis=1)
```

```python
train=train.drop(columns=['Location','Name'],axis=1)
```

```python
train
```

```python
col=train.pop('Price')
```

```python
train
```

```python
train['Price']=col
```

```python
train
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(train.iloc[:, :-1],train.iloc[:, -1], test_size =0.3,random_state = 50)
```

```python
model =RandomForestRegressor()
```

```python
model.fit(Xtrain,Ytrain)
```

```python
Ypredict=model.predict(Xtest)
```

```python
r2_score(Ytest,Ypredict)*100
```

```python
#Accuracy is approximately 90% with r2_score
```

```python
mean_squared_error(Ytest,Ypredict)
```
